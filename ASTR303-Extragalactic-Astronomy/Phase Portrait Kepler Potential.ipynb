```python
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.mlab as mlab
import math
%pylab inline
```

```python
k = 3
m = 1
mu = m/2
r = linspace(1,5.5,40)
l = np.array([0,0.5,1,1.5,2])
print(r)
print(l)
```

```python
def Ueff(l,r):
    Ueff = -k/r + (l**2)/(2*mu*(r**2))
    return Ueff
```

```python
plt.figure(figsize=(8,8))
for n in l:
    
    plot(r,Ueff(n,r))
   
    grid(1)
    xlabel("r")
    ylabel("U$_eff$")
    title('Effective Potential')
    legend(['l=0','l=0.5','l=1','l=1.5','l-2'], loc=5)
    plt.tight_layout()
    
```

```python
plt.figure(figsize=(8,8))
for n in l:
    plt.quiver(r,Ueff(n,r))
```

```python
from sympy.interactive import printing # https://www.youtube.com/watch?v=Dkifb6nytao
printing.init_printing(use_latex=True)
```

```python
M = sp.Symbol('M') # point mass
G = 6.67e-11 # gravitational constant
r = sp.Symbol('r') #separation

t = sp.Symbol('t')
x = sp.Function('x')(t)
v = sp.Function('v')(x) #velocity

diffeq = Eq(x.diff(t,t), -G*M/r**2)
display(diffeq)

dsolve(diffeq,x) 
```

```python
a = Eq(v.diff(x),-G*M/r**2)
v = Eq(x.diff(t),v)

display(a)
display(v)

dsolve(a,x)
```
