```python
%pylab nbagg
%matplotlib inline
import sympy as sp
import numpy as np
```

```python
## DISK n = 1
## -------------

r = sp.Symbol('r') #Radius 
reff = sp.Symbol('reff') #scalelength

c = r/reff
c= linspace(0,3,50)

n = 1 #index
b = 5/3 #b n=1

Σ = sp.Function('Σ')(r) 
Σ = sp.exp(-b*((c**(1/n))-1)) #Sérsic Func.
Σl = (-b*((c**(1/n))-1)) # ln(Σ)
```

```python
## ELLIPTICAL n = 4
## -------------

i = 4
bf = 23/3
Λ = sp.exp(-bf*((c**(1/i))-1)) #Sérsic Func. n = 4 // de Vacouleur's law
Λl = (-bf*((c**(1/i))-1))
```

```python
for i in c:
    g=array(Λl,dtype=float)
    G = g.size

    plt.grid(True)
    plt.plot(g,c, 'b-')
    xlabel("R/R$_eff$")
    ylabel("ln$Σ(r)$")
    title('Σ(r) for an Elliptical $n=4$')
    legend(['Σ$_n=4$ (elliptical)'], loc=5)
    plt.tight_layout()
```
