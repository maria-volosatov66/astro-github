```python
freq = array([1e2,5e2,1e3,2e3,5e3,10e3,100e3])


vin = array([7.25,7.3,7.27,7.31,7.43,7.57,7.66])


vout= array([7.03,7.06,7.00,6.91,6.29,4.82,0.3])


phase = -array([0.92,3.59,6.93,13.1,29.9,43.98,98.3]) *(pi/180)

G =20*log10(vout/vin)

figure(figsize=[15,6])
ff = logspace(1,6,100)

R=978
L=18e-3
z2=R
z1=1j*2*pi*ff*L
Ttheo = z2/(z1+z2)

Gtheo = 20*log10(abs(Ttheo))
phase_theo = angle(Ttheo)*180/pi

subplot(1,2,1)
semilogx(ff,Gtheo,freq,G,'o')
legend(['Theory','Experiment'], loc=4)
grid(1)
ylim([-35,3])
xlabel('Frequency [Hz]')
ylabel('Gain [dB]')
title('Bode Plot of Low Pass RL Filter-Amplitude Transfer Function')

subplot(1,2,2)
semilogx(ff,angle(Ttheo)*180/pi,freq,phase*180/pi,'o')
grid(True)
legend(['Theory','Experiment'])
xlabel('Frequency [Hz]')
ylabel('Phase[deg]')
title('Phase Shift of Low Pass RL Filter')
```
