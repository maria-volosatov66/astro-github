```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
f = np.array([59,100.8,149.7,200.0,250.0,304.9,352.1,403.2,454.4,497.5,549.5,602.4])/2
x = np.array([0.02,0.32,0.36,0.51,0.76,1.07,1.15,1.35,1.68,1.84,1.92,2.29])/10/100
dx = 1e-4
df = 1
```

```python
slope1, intercept1 = np.polyfit(f,x, 1)
print("Slope =",slope1, "Intercept = ", intercept1)
```

```python
plt.plot(f,x, 'ro')
plt.errorbar(f, x, yerr=dx, fmt = 'o', color='blue')

plt.xlabel('Frequency f (Hz)')
plt.ylabel('Displacement x (m)')
plt.title('Displacement vs Frequency - Method I Speed of Light')
plt.grid(True)
print("Slope =",slope1, "Intercept = ", intercept1)
```

```python
D = np.array([15.78,22.61,49.69])
t = np.array([100,148,320])*10**-9
print(D)
```

```python
import statistics as stat
```

```python
sig= stat.stdev(D/t)
print(sig)
```

```python
slope2, intercept2 = np.polyfit(t,D, 1)
print("Slope =",slope2, "Intercept = ", intercept2)
y = slope2*t+intercept2
```

```python
plt.plot(t,D, 'mo')
plt.xlabel('Time t (s)')
plt.plot(t,y, 'c--')
plt.legend(['Data','Fit'])
plt.ylabel('Displacement x (m)')
plt.title('Displacement vs Time - Method II Speed of Light')
plt.grid(True)
print("Slope =",slope2, "Intercept = ", intercept2)
```
