```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
a = 2e-6
b = 10e-6
λ = 6.32e-7
z = 1
```

```python
N = 100
x = np.linspace(-1,1,N)
X, Y = np.meshgrid(x,x)
R = np.sqrt(X**2+Y**2+z**2)
kx = (2*np.pi*x)/(λ*z)
ky = (2*np.pi*Y)/(λ*z)
In = (2*np.pi)/(λ**2)
```

```python
ß = ky*b/(2*z)
å = kx*a/(2*z)
```

```python
I = (a*a*b*b)*((np.sinc(a*kx*x/2)**2))*((np.sinc(b*ky*Y/2)**2))
II = a*a*b*b*(np.sin(ß)/ß)*(np.sin(å)/å)/λ**2
Ix = (a**2)*(b**2)*((np.sinc(a*kx/2))**2)*(np.sinc(b*ky/2)**2)
```

```python
plt.figure(figsize=(7,7))
plt.imshow(II, extent=[-100,100,-100,100], cmap='magma')
plt.title('Rectangular Aperture')
```
