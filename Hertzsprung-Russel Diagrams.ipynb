```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

## Importing numpy for manipulating arrays
## Importing matplotlib for plotting isochrones for data
```

```python
log_age, log_lum, log_Teff = np.loadtxt('isoc_z008.dat',usecols = (0,3,4),unpack=True)

## np.loadtxt() <--for opening data/plain text file + reading!
### usecols == specifies which columns to read

## 0 == column for log of age 
## 3 == log (L/L*) luminosity
## 4 == log (T eff) == effective temperature
```

```python

w7 = np.where(log_age == 7.0) #array for log(age) == 7
w8 = np.where(log_age == 8.0) #array for log(age) == 8

plt.plot(log_Teff[w7], log_lum[w7],label='10 Myr')
plt.plot(log_Teff[w8], log_lum[w8],label='100 Myr')

plt.xlabel('Log(Temperature $_e$$_f$$_f$)')
plt.ylabel('Log Luminosity')
plt.axis([5.0, 3.5, 0, 6])
plt.legend(loc='lower left')
plt.show()
```

```python
w75 = np.where(log_age == 7.5) #array for log(age) == 7.5
w85 = np.where(log_age == 8.5) #array for log(age) == 8.5
w9 = np.where(log_age == 9.0) #array for log(age) == 9.0

plt.plot(log_Teff[w75], log_lum[w75],label='10 Myr')
plt.plot(log_Teff[w85], log_lum[w85],label='100 Myr')

plt.xlabel('Log(Temperature $_e$$_f$$_f$)')
plt.ylabel('Log Luminosity')
plt.axis([5.0, 3.5, 0, 6])
plt.legend(loc='lower left')
plt.show()
```

### 1.3
Plot a single Hertzsprung-Russell diagram in which you show the isochrones for
log(age) = [7.0, 7.5, 8.0, 8.5, 9.0] for metallicity Z = 0.008.

```python
plt.figure(figsize=(7,7))



plt.plot(log_Teff[w7], log_lum[w7],label='log_age = 7.0')
plt.plot(log_Teff[w75], log_lum[w75],label='log_age = 7.5')
plt.plot(log_Teff[w8], log_lum[w8],label='log_age = 8.0')
plt.plot(log_Teff[w85], log_lum[w85],label='log_age = 8.5')
plt.plot(log_Teff[w9], log_lum[w9],label='log_age = 9.0')

plt.title('Isochrones for Z=0.008')
plt.grid(True)
plt.axis([4.5, 3.5, -3, 6])
plt.xlabel('log(Temperature)')
plt.ylabel('log(Luminosity)')
plt.legend(loc='lower right')

plt.savefig('HRz008.pdf')
```
