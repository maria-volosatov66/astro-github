#### CALCULATING REQUIRED EXPOSURE TIME FOR S/N VALUES

```python
Flux_20 = 10**((-0.3597002995010542*20)+8.733143150633994)
Flux_20_Du = Flux_20/0.435**2
print('Flux of a 20 magnitude star = ',Flux_20)
print('Flux of a 20 magnitude star in DU/arcsec^2 = ',Flux_20_Du)
g = 3 #gain
sigma_RO = 1.54 # Readout noise
A = np.pi*15**2 # area of aperture used
SN = np.array([10,20,50,100]) 
a = Flux_20*g/2
b = (msky_mean_du*0.435**2)*g*A/2
c = (sigma_RO*g)**2*A
ad = Flux_20*g/2
bd = (msky_dark*0.435**2)*g*A/2
cd = ((sigma_RO**2)*g)**2*A
print('Flux*gain/2 =',a)
print('msky*g*A/2 =',b)
print('(Sigma*g)^2*A =', c)
```

```python
# rearranging, S/N = at / (at+bt+c)^1/2
# By having a value for S/N, we can solve for t as it is an ordinary quadratic eqn
for i in SN:
    A = a**2
    B = -(SN**2)*(a+b)
    C = -c*(SN**2)
print('A =',A)
print('B =',B)
print('C =',C)
```

```python
#FOR DARK SKY RESERVE USING MSKY = .46
for i in SN:
    AD = ad**2
    BD = -(SN**2)*(ad+bd)
    CD = -cd*(SN**2)
print('AD =',AD)
print('BD =',BD)
print('CD =',CD)
```

#### The above calculations depend on both the star aperture and sky annulus sizes. Explain why.

The aperture is optimal if it captures most of the light from the star but includes minimal of the background. Limiting magnitude is dependant on the instrument's collecting power, which in aperture photometry depends on the summed up pixel values within a circular region around the star, and the annulus surrounding it to account for the light from the sky. For a faint star, a large aperture (and large annulus size) can, depending on sky brightness, drown out the star itself, and information will be lost, and if there are many faint stars in the sky, the limiting magnitude may be brighter than expected because only stars which are much brighter than the others could feasibly be accounted for. On the other hand, choosing an aparture which does not encompass all of the starlight or does not include much of the background will cause loss of information and background calibration, and produce a calculation for the limiting magnitude that is inaccurate. The above calculation's results is as good as the calibration equation determined from the Log(flux) vs log (Rmag) plot; improvements can be made by considering more stars and/or choosing a different apperture.

#### Explain what this noise (scintillation) would do to your S/N vs. S plot. 
Scintillation noise variability is largely due to changes in the optical turbulence profile, which occurs due to mechanical mixing of layers of air with a temperature and density gradient. The locally focused and defocused incoming wavefront results in a array of spatial intensity fluctuations or a speckle pattern. Wind speed also affects the measure of scintillation noise as it causes the speed of the variations, and in turn causes variations in noise that limit photometric measurements. 
    The scintillation index is given by Ïƒ^2=<I^2> âˆ’ < I >^2 / < I >^2   [1]
    <br />
 For short exposures, the scintillation index can also be calculated using the parameters of the telescope and atmospheric conditions (Darvins et al. 1998)[2] Ïƒ^2 = 17.34D^-7/3 (cos Î³)^-3 Integral(0,âˆž)(h^2C^2(h) dh), where D is the telescope diameter,Cn(h) is the refractive index structure constant that measures the strenght of turbulence, h is the altitude of the turbulence layer, and the cosine term gives the azimuthal angle of the star. For longer exposures, exposure time is also a dependant factor. The scintillation rms noise is significantly reduced for fainter stars as the photon noise becomes equivalent to the scintillation noise at a certain limiting magniitude, which depends on telescope parameters, atmospheric conditions, and the zenith angle. 
    As the scintillation's power is proportional to the flux, the scatter will increase with increasing photon noise and converge to some constant value, while very faint stars are subject to random fluctuations. [3]
<br />
Scintillation noise will prevent us from seeing the Poisson noise regions in the S/N vs S plots since the the noise would dominate at 2 seconds of exposure; this is mostly visible for the sky S/N vs S plot.


#### Explain why errors due to not flat fielding, or to non-photometric observing conditions, appear in the S/N--S plot in a similar fashion to scintillation noise. 

Errors due to not flat-fielding do not remove a floating bias from the frames on which photometry is done; since random noise (Poisson noise and readout noise and dark current) are not accounted for in the original frames, the flux of the stars on which photometry is done will be prone to the same errors as non-photometric analysis, and since N won't be reduced, regions where the signal is too close to the noise will result in regions in the plot where certain regiemes of noise cannot be detected--in a manner similar to scintillation noise since that also requires correction and with greater scintillation, information is not preserved well for greater magnitudes.
